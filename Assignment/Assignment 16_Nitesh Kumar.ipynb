# Q2.What are Data Collection Method in Recommender System. 

### Answer:

Data collection is a crucial step in building and maintaining recommender systems, as the quality and quantity of data directly impact the system's effectiveness. There are several data collection methods and sources used in recommender systems:

#### 1. User Interactions:

##### Clickstream Data:
Collecting data on user interactions, such as clicks, views, purchases, and likes, to understand user preferences.

##### Ratings and Reviews:
Gathering user-provided ratings and reviews for products or content to assess user preferences and sentiments.

#### 2. Implicit Feedback:

Implicit user feedback, like dwell time (how long a user stays on a page or interacts with an item), can provide valuable information about user interests.

#### 3. User Profiles:

##### User Registration Data:
Information collected during user registration, including demographics, location, and user-generated profiles.

##### User Surveys:
Conducting surveys to gather explicit user preferences, interests, and demographic information.

##### User Activity History:
Tracking user behavior and actions over time to build user profiles.

#### 4. Content Data:

##### Item Metadata:
Collecting information about the items themselves, such as titles, descriptions, categories, and tags.
Item Attributes: Incorporating additional attributes specific to items, such as actors in a movie or ingredients in a recipe.

#### 5. Contextual Data:

Collecting contextual information like the time of day, location, or device used, which can influence user preferences and recommendations.

#### 6. Social Network Data:

Utilizing data from social networks to understand relationships and connections between users, which can be used for collaborative filtering or social recommendations.

#### 7. Crowdsourced Data:

Collecting data from external sources or user-generated content, such as data from APIs, review websites, or crowdsourcing platforms.

#### 8. A/B Testing:

Running controlled experiments to collect data on user responses to different recommendation algorithms or strategies.

#### 9. Data Scraping:

Automatically extracting data from websites, forums, or online communities to gather information about products, services, or user opinions.

#### 10. Mobile and App Data:

Collecting data from mobile applications or web apps, including location data, in-app behaviors, and user preferences.

#### 11. Sensor Data:

Gathering data from sensors, IoT devices, or wearables to understand user behavior and preferences in the physical world.

#### 12. In-App Surveys:

Prompting users within the application to provide feedback on their preferences, interests, or user experience.

#### 13. Purchase History:

Collecting data on user purchase history, which is valuable for e-commerce recommender systems.

Data collection methods should be aligned with the goals of the recommender system and respect user privacy and data protection regulations. Additionally, data should be cleaned, preprocessed, and anonymized as needed to ensure data quality and user privacy.



# Q3. Build a Basic Recommender system

### Answer:

We will build Movie recommendation system

```python
# import library
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
credits.head()
```

```python
movies_df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies_df.head()
```

```python
print("Credits:",credits.shape)
print("Movies Dataframe:",movies_df.shape)
```

```python
# In credit dataset we renamed the movie_id to id. Because movie_df have id columns already.
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})
```

```python
# merge that id column with movie_df id column
movies_df_merge = movies_df.merge(credits_column_renamed, on='id')
# check top 5 data
movies_df_merge.head()
```

```python
# drop irrelevent features and store in new dataframe
movies_cleaned_df = movies_df_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
# check top 5 data
movies_cleaned_df.head()
```

```python
# information of our cleaned data
movies_cleaned_df.info()
```

### Formula for recoomendation system for guest user.
In guest user history of user not record.

```python
# Calculate all the components based on the above formula
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)
```

### recomendation system based on weighted average

```python
# create new column of weighted average and calculate the value by using formula given above
movies_cleaned_df['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
# check top 5 data
movies_cleaned_df.head()
```

```python
# sorted value of weighted average column
movie_sorted_ranking=movies_cleaned_df.sort_values('weighted_average',ascending=False)

# check particular column that seems to be more important
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(20)
```

```python
weight_average=movie_sorted_ranking.sort_values('weighted_average',ascending=False)
```

```python
weight_average
```

```python
# visualization
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10),
                  y=weight_average['original_title'].head(10), data=weight_average)

plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

### recomendation system based on popularity based

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10),
               data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

```python
popularity.head()
```

### Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_cleaned_df[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_average']*0.5+movies_cleaned_df['normalized_popularity']*0.5
```

```python
movies_scored_df = movies_cleaned_df.sort_values(['score'], ascending=False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = movies_cleaned_df.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

plt.savefig('scored_movies.png')
```
