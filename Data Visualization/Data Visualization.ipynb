```python
#Create line chart --> plot(x,y)
x=range(1,11) #x axis
y=[1,5,8,9,2,0,3,10,4,7] #y axis
plt.plot(x,y) # range is x axis ie pts(1,1),(2,5),(3,8)..last x=10
#Can Change the scale
plt.xlim(0,20) #Xscale 0,20 --> Limit Range
plt.ylim(0,20) #Yscale 0,20 --> Limit Range
plt.show()
```

```python
#Create Multiple line chart --> plot(x,y)
x=range(1,11) #x axis
y=[1,5,8,9,2,0,3,10,4,7] #y axis
y1=[2,4,6,8,1,3,5,7,9,2]
plt.plot(x,y) # range is x axis ie pts(1,1),(2,5),(3,8)..last x=10
plt.plot(x,y1)
#Can Change the scale
#plt.xlim(0,20) #Xscale 0,20 --> Limit Range
#plt.ylim(0,20) #Yscale 0,20 --> Limit Range
plt.savefig('Multiple Line Chart.png')
plt.show()
```

```python
# Create Pie Chart, plots the % of each value
values=[5,8,9,10,4,7]
#color
col=['b','g','c','m','y','r']
#label
lab=['A','B','C','D','E','F']
# Cut Slice (Explode)
exp=[0,0,0,0,0,0.9]

plt.pie(values,colors=col,labels=lab,explode=exp,
        autopct='%.f%%',counterclock=False) #autopct --> string formatting for display of percentage
plt.show()
```

```python
#Pie Chart
country=['Norway','Germany','Canada','United States','Netherlands']
medal=[14,15,11,9,8]
col=['b','g','m','c','y']
exp=[0,0.2,0,0,0]
plt.pie(medal,colors=col,labels=country,explode=exp,autopct='%1.2f%%',shadow=True)
plt.title('Medal Tally',fontsize=15)
plt.show()
```

```python
#Create Bar Chart
x=range(1,7)
y=[5,8,9,10,4,7]
col=['b','g','c','m','y','r']
plt.bar(x,y,color=col)
plt.savefig('Bar Chart.png') #Save fig before plt.show()
plt.show()
```

A density plot is a representation of the distribution of a numeric variable. It uses a kernel density estimate to show the probability density function of the variable (see more). It is a smoothed version of the histogram and is used in the same concept.

Density Plot is a type of data visualization tool. It is a variation of the histogram that uses 'kernel smoothing' while plotting the values. It is a continuous and smooth version of a histogram inferred from a data. ... Density plots can be made using pandas, seaborn, etc.

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable.

```python

sns.kdeplot(cars.horsepower)

#Most Cars have the Horse power of 80, some cars have around 150
```

```python
sns.kdeplot(cars.horsepower, shade =True)
```

```python
#Bin Width (bw) makes a lot of difference in the kde plot, if bw is increased or decreased the plot changes. Peaks appear
sns.kdeplot(cars.horsepower, bw=10) 
```

```python
sns.kdeplot(cars.horsepower,bw=100) 
#As we increase the bin width the curve smoothens and tends to be normal #Looses out on peaks
```

```python
#Cumulative Density Plot (cdf)
sns.kdeplot(cars.horsepower,cumulative=True)
```
